# Add fixed obstacles to a driving environment

```julia
using AutomotiveDrivingModels
using AutoViz
using AutomotivePOMDPs
```

```julia
cam = FitToContentCamera(0.)
rng = MersenneTwister(1);
```

## Urban Environment (T intersection + crosswalks)

```julia
params = UrbanParams(nlanes_main=1,
                     crosswalk_pos =[VecSE2(6, 0., pi/2), VecSE2(-6, 0., pi/2), VecSE2(0., -5., 0.)],
                     crosswalk_length =  [14.0, 14., 14.0],
                     crosswalk_width = [4.0, 4.0, 3.1],
                     stop_line = 22.0)
env = UrbanEnv(params=params);
```

The environment has no obstacles by default

```julia
render(Scene(), env, cam=cam)
```

initialize an obstacle distribution

```julia
obs_dist = ObstacleDistribution(env, upper_obs_pres_prob=0., left_obs_pres_prob=1.0, right_obs_pres_prob=1.0);
```

sample obstacles from it and add them to the environment:

```julia
sample_obstacle!(env, obs_dist, rng)
render(Scene(), env, cam=cam)
```
